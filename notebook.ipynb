## Loading Data from Pandas DataFrame

```python
import pandas as pd
```

```python
pandas_df = pd.read_csv('titanic.csv')
df = ge.from_pandas(pandas_df)
df.head()
```

## Loading Data from Spark DataFrame

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
    .master('local[1]') \
    .appName('GreatExpectations') \
    .getOrCreate()
```

```python
spark_df = spark.read \
    .option('header', 'true') \
    .csv('titanic.csv')
spark_df.show()
```

```python
from great_expectations.dataset.sparkdf_dataset import SparkDFDataset
```

```python
df = SparkDFDataset(spark_df)
df.head()
```

## Adding Expectations 

### Expecting Values in Set
